```python
train=ImageDataGenerator(rescale=1/255)
validation=ImageDataGenerator(rescale=1/255)
```

```python
train_dataset=train.flow_from_directory("C:\\Users\\chait\\OneDrive\\Desktop\\Face_detection_cnn\\train",
                                        target_size=(200,200),
                                       batch_size=5,
                                       class_mode='binary')
validation_dataset=train.flow_from_directory("C:\\Users\\chait\\OneDrive\\Desktop\\Face_detection_cnn\\validation",
                                        target_size=(200,200),
                                       batch_size=5,
                                       class_mode='binary')
```

```python
train_dataset.class_indices
```

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16,(3,3),activation="relu",input_shape=(200,200,3)),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Conv2D(32,(3,3),activation="relu",input_shape=(200,200,3)),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Conv2D(64,(3,3),activation="relu",input_shape=(200,200,3)),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(64,activation='relu'),
    tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model_fit=model.fit(train_dataset,
                    epochs=10,
                    validation_data=validation_dataset)
```

```python
dir_path="C:\\Users\\chait\\OneDrive\\Desktop\\Face_detection_cnn\\test"
import numpy as np

for i in os.listdir(dir_path):
    img = image.load_img(path=dir_path + '\\' + i, target_size=(200, 200))
    

    X=image.img_to_array(img)
    X=np.expand_dims(X,axis=0)
    images=np.vstack([X])
    val=model.predict(images)
    if val==0:
        print("the image is of a man")
    else:
        print("the image is of a women")
```
